```python
image_train = '/content/data_yolo_format/images/train'
for filename in os.listdir(image_train):
  name = os.path.basename(filename)
  f_name, f_ext = os.path.splitext(name)
  outputFileName = f_name + ".txt"
  # Walking top-down from the root
  for root, dir, files in os.walk('/content/data/labels'):
      if outputFileName in files:
        src = os.path.join('/content/data/labels',outputFileName)
        dst = '/content/data_yolo_format/labels/train'
        shutil.copy(src, dst)

image_val = '/content/data_yolo_format/images/val'
for filename in os.listdir(image_val):
  name = os.path.basename(filename)
  f_name, f_ext = os.path.splitext(name)
  outputFileName = f_name + ".txt"
  # Walking top-down from the root
  for root, dir, files in os.walk('/content/data/labels'):
      if outputFileName in files:
        src = os.path.join('/content/data/labels',outputFileName)
        dst = '/content/data_yolo_format/labels/val'
        shutil.copy(src, dst)

image_test = '/content/data_yolo_format/images/test'
for filename in os.listdir(image_test):
  name = os.path.basename(filename)
  f_name, f_ext = os.path.splitext(name)
  outputFileName = f_name + ".txt"
  # Walking top-down from the root
  for root, dir, files in os.walk('/content/data/labels'):
      if outputFileName in files:
        src = os.path.join('/content/data/labels',outputFileName)
        dst = '/content/data_yolo_format/labels/test'
        shutil.copy(src, dst)
```
