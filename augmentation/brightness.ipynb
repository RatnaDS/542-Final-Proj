```python
image = cv2.imread('/content/data/images/case_00000_00091.png')
```

```python
new_image = np.zeros(image.shape, image.dtype)
# alpha = 1.0 # Simple contrast control
# beta = 0    # Simple brightness control

alphavalues = [1.2,1.3,1.4]
betavalues = [20,30,40]

alpha = random.choice(alphavalues)
beta = random.choice(betavalues)

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        for c in range(image.shape[2]):
            new_image[i,j,c] = np.clip(alpha*image[i,j,c] + beta, 0, 255)
```
